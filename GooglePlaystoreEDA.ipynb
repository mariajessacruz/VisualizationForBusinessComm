```python
df['Reviews']=df['Reviews'].str.replace('3.0M', '3000000')
```

```python
# Size Column
df['Size'].unique()
```

```python
def convert_size(size):
  if 'M' in size:
        return int(float(size.replace('M', ''))*1000)
  elif 'k' in size:
    return float(size.replace('k', ''))
  elif 'Varies with device' in size:
    return np.nan
  else:
    return np.nan
```

```python
df['Size']=df['Size'].apply(convert_size)
```

```python
df.info()
```

```python
df['Installs'].unique()
# problem: '+', ','
```

```python
df['Price'].unique()
# problem: '$'
```

```python
# method 1
cols=['Installs', 'Price']
chars_to_remove=['+', ',', '$']
for i in cols:
  for j in chars_to_remove:
    df[i]=df[i].str.replace(j,'')
```

```python
# method 2
df['Installs']=df['Installs'].str.replace(',','')
df['Installs']=df['Installs'].str.replace('+','')
df['Price']=df['Price'].str.replace('$','')
```

```python
df.info()
```

```python
df['Installs']=df['Installs'].astype('int')
df['Price']=df['Price'].astype('float')
```

```python
df.info()
```

```python
df['Last Updated']=pd.to_datetime(df['Last Updated'])
df['day']=df['Last Updated'].dt.day
df['month']=df['Last Updated'].dt.month
df['year']=df['Last Updated'].dt.year
```

```python
# Remove duplicates from my column app
df.drop_duplicates(subset=['App'],keep='last',inplace=True)
df
```

**Q11. Are there any missing values in the dataset? Identify any missing values and describe how they may impact your analysis.**

```python
# Missing Values
def impute_median(series):
    return series.fillna(series.median())

df['Rating'] = df['Rating'].transform(impute_median)
```

**Q12. What is the relationship between the size of an app and its rating? Create a scatter plot to visualize the relationship.**

```python
# Create a scatter plot function
def scatters(col1, col2):
    # Create a scatter plot using Seaborn
    plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
    sns.scatterplot(data=df, x=col1, y=col2, hue="Type")
    plt.title(f'Scatter Plot of {col1} vs {col2}')
    plt.xlabel(col1)
    plt.ylabel(col2)
    plt.show()
```

```python
scatters('Size', 'Rating')
```

**Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.**

```python
# Type of app by average price
genres = df.groupby('Genres')['Price'].mean().sort_values(ascending=False)

# Create a grouped bar chart
plt.figure(figsize=(30, 28))
plt.title('Average Prices of Apps by Genres')
plt.xlabel('Genre')
plt.ylabel('Average Price')

# Create the grouped bar chart
genres.plot(kind='bar', color='skyblue')

# Show the chart
plt.show()
```

**Q16. A mobile app development company wants to analyze the Google Playstore dataset to identify the most successful app developers. What features would you analyze to make recommendations to the company, and what data visualizations would you use to present your findings?**

```python
# Considering the number of installs by category Analysis
installs_by_category = df.groupby('Category')['Installs'].sum().sort_values(ascending=False).head(10)
```

```python
# number of installs by category
plt.figure(figsize=(12, 6))
sns.barplot(x=installs_by_category.index, y=installs_by_category.values, hue=installs_by_category.index, palette='viridis', legend=False)
plt.title('Total Number of Installs by Top 10 Categories')
plt.xlabel('Category')
plt.ylabel('Total Installs')
plt.xticks(rotation=45)
plt.show()
```

```python
# Average Rating Analysis
avg_rating_by_category = df.groupby('Category')['Rating'].mean().dropna().sort_values(ascending=False).head(10)
```

```python
# average rating by category visualization
plt.figure(figsize=(12, 6))
sns.barplot(x=avg_rating_by_category.index, y=avg_rating_by_category.values, hue=avg_rating_by_category.index, palette='viridis', legend=False)
plt.title('Average Rating by Top 10 Categories')
plt.xlabel('Category')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
# Number of Reviews Analysis
reviews_by_category = df.groupby('Category')['Reviews'].sum().sort_values(ascending=False).head(10)
```

```python
# number of reviews by category visualization
plt.figure(figsize=(12, 6))
sns.barplot(x=reviews_by_category.index, y=reviews_by_category.values, hue=reviews_by_category.index, palette='rainbow', legend=False)
plt.title('Total Number of Reviews by Top 10 Categories')
plt.xlabel('Category')
plt.ylabel('Total Reviews')
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribution of content ratings visualization
plt.figure(figsize=(8, 8))
content_rating_counts = df['Content Rating'].value_counts()
plt.pie(content_rating_counts, labels=content_rating_counts.index, autopct='%1.1f%%', colors=sns.color_palette('viridis', len(content_rating_counts)))
plt.title('Distribution of Content Ratings')
plt.show()
```

```python
# Convert 'Last Updated' column to datetime
df['Last Updated'] = pd.to_datetime(df['Last Updated'])
```

```python
# Calculate time since last update for each app
df['Days Since Last Update'] = (pd.Timestamp.now() - df['Last Updated']).dt.days
```

```python
# average days since last update by category
avg_update_frequency_by_category = df.groupby('Category')['Days Since Last Update'].mean().sort_values(ascending=True).head(10)
```

```python
# average update frequency by category visualization
plt.figure(figsize=(12, 6))
sns.barplot(x=avg_update_frequency_by_category.index, y=avg_update_frequency_by_category.values, hue=avg_update_frequency_by_category.index, palette='viridis', legend=False)
plt.title('Average Update Frequency by Top 10 Categories')
plt.xlabel('Category')
plt.ylabel('Average Days Since Last Update')
plt.xticks(rotation=45)
plt.show()
```
